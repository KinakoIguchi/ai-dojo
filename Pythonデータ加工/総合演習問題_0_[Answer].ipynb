#### 【Question1】果物価格の平均を算出する
---

'testdata1.csv'は果物と価格の入ったデータです。  
データを読み込んで確認の為、データ表示した上で、価格の平均を算出するコードを書きました。  
しかしエラーとなってしまいました。  

以下のコードを修正し、どのコードをなぜ修正したか纏めてみましょう。

````python
# ライブラリインポート
import pandas as pd

test1 = pd.read_csv('data/testdata1.csv')

# データの表示
print(test1)

# 平均価格表示
print(果物の平均価格:, test1['価格'].mean())
````

●修正箇所と修正理由● ※どの箇所をどう修正したかとなぜ修正したかを簡単に整理して記載してみましょう（自由記述）

####  以下解答

````python
# ANSWER
# ライブラリインポート
import pandas as pd

test1 = pd.read_csv('data/testdata1.csv', encoding='cp932')

# データの表示
print(test1)

test1['価格'] = test1['価格'].str.replace('円', '')
test1['価格'] = test1['価格'].str.replace(',', '').astype(int)

# 平均価格表示
print('果物の平均価格:', test1['価格'].mean())
````

#### 【Question2】現データを残しつつマスタデータの更新をする
---

'testdata2.csv'はある文房具店の商品と価格の入ったマスタデータです。  
来月から販売する商品に以下の変更が入る予定です。  

* 「研磨式消しゴム」の価格を高すぎと判断して500円 -> 450円に変更
* 「ただの輪ゴム」の商品名がタダじゃないとクレームがあった為、「何の変哲もない輪ゴム」に変更
* 「片手で書けるえんぴつ」は当たり前すぎて売れない為、販売終了
* 以前人気だった「曲がらないコンパス」を復刻版として新たに200円で販売

来月に向けてマスタデータを事前に変更することにしました。  
現商品マスタは現在使用中かつ過去分を残しておきたいので、直接修正は行わず一旦新しいデータフレームで修正し、  
来月になったら入れ替えを行うようにします。  
  
修正を行ったところ、現在使用しているマスタデータにも一部のみ変更が反映されてしまいました。  


以下のコードを修正し、どのコードをなぜ修正したか纏めてみましょう。

````python
# ライブラリインポート
import pandas as pd
test2 = pd.read_csv('data/testdata2.csv', index_col=0)

# 読み込んだデータの表示
print(test2)

test2_new = test2

# 価格・商品名の変更
test2_new.loc[test2_new['商品名']=='研磨式消しゴム', '価格'] = '450円'
test2_new.loc[test2_new['商品名']=='ただの輪ゴム', '商品名'] = '何の変哲もない輪ゴム'

# 販売終了商品の削除
test2_new = test2_new.drop(index='S0073')

# 新商品の追加
newitem = pd.Series(['曲がらないコンパス', '200円'], index=test2_new.columns, name='S0003')
test2_new = test2_new.append(newitem)

# 修正の確認
print('')
print('=新商品マスタ==========')
print(test2_new)
print('')
print('=現商品マスタ==========')
print(test2)
````

●修正箇所と修正理由● ※なぜ想定外な結果となったかやどの箇所をどう修正したか等を簡単に整理して記載してみましょう（自由記述）  
* XXXX


####  以下解答

````python
# ANSWER
import pandas as pd
test2 = pd.read_csv('data/testdata2.csv', index_col=0)

# 読み込んだデータの表示
print(test2)

test2_new = test2.copy()

# 価格・商品名の変更
test2_new.loc[test2_new['商品名']=='研磨式消しゴム', '価格'] = '450円'
test2_new.loc[test2_new['商品名']=='ただの輪ゴム', '商品名'] = '何の変哲もない輪ゴム'

# 販売終了商品の削除
test2_new = test2_new.drop(index='S0073')

# 新商品の追加
newitem = pd.Series(['曲がらないコンパス', '200円'], index=test2_new.columns, name='S0003')
test2_new = test2_new.append(newitem)

# 修正の確認
print('')
print('=新商品マスタ==========')
print(test2_new)
print('')
print('=現商品マスタ==========')
print(test2)
````

【ANSWER】  
●修正箇所と修正理由● ※なぜ想定外な結果となったかやどの箇所をどう修正したか等を簡単に整理して記載してみましょう（自由記述）  
* Pythonで変数を定義(CSVからDataFrameを作成等)したとき、変数にはobjectIDが格納される。
* df2 = dfとした場合、df2にはdfに格納されたobjectIDが代入されるため、objectを参照する。
* そのためdf2に変更を加えるとobject参照しているdfも変更される。
  
  
* df2 = df.copy()とするとobjectを複製したものがdf2に代入されるため、objectIDが変わる。
* そのためdf2に変更を加えてもdfは変更されない。

```
# 以下確認のためのコード
df_1 = pd.DataFrame({'a':[1,2,3,4],'b':[50,100,200,400]})
df_2 = df_1
print('=だとidが変わらない(参照渡し)')
print(f'df_1:{id(df_1)} df_2:{id(df_2)}')
df_3 = df_1.copy()
print('copyだとidが変わる')
print(f'df_1:{id(df_1)} df_3:{id(df_3)}')
```

#### 【Question3】野球選手の年俸の差額を確認する
---

'testdata3_1.csv'と'testdata3_2.csv'は、ヤクルトの野球選手のデータです。  

* 'testdata3_1.csv'：2012年の選手の属性情報と年俸および打率などの成績のデータ   
* 'testdata3_2.csv'：2013年の選手の年俸のデータ 

野球選手の年俸について分析することにしました。  
最初に2012年から2013年でどれくらい年俸に差があるのか差額の平均を算出しようと思います。  

データが2ファイルに分かれているので、まずデータを結合する必要があります。  
背番号は異なる選手で同じ背番号が存在しているので、選手名で結合することにしました。  
また年俸の差額を見たいので、2012年にも2013年にも在籍している選手のみを算出対象にします。  

実際に結合を行ったところ、結合後データのデータ数が2つの元データより増えてしまいました。  
ヤクルトで2012年と2013年の両方で在籍していた選手は**31人**のはずです。  

以下のコードを修正し、どのコードをなぜ修正したか纏めてみましょう。  
また今回は原因を探すための探索過程のコードも消さずに残しておきましょう。

````python
# ライブラリインポート
import pandas as pd
test3_1 = pd.read_csv('data/testdata3_1.csv')
test3_2 = pd.read_csv('data/testdata3_2.csv')
````

````python
# 読み込んだデータの表示
test3_1.head()
````

````python
# 読み込んだデータの表示
test3_2.head()
````

````python
# 結合前件数確認
print('test3_1件数:', test3_1['選手名'].count())
print('test3_2件数:', test3_2['選手名'].count())
````

````python
# データ結合
test3_merge = pd.merge(test3_1, test3_2, on='選手名', how='inner')

# 年俸の差分を算出（2013年年俸 - 2012年年俸）
test3_merge['年俸差分'] = test3_merge['年俸_y'] - test3_merge['年俸_x']

# 年俸の差分の平均を算出
print(f"平均年俸差分：{test3_merge['年俸差分'].mean()} 万円")
````

````python
# 結合後件数確認
print('結合後件数:', test3_merge['選手名'].count())
````

####  以下解答

````python
# ANSWER
# 結合前件数確認
print('test3_1件数:', test3_1['選手名'].count(), '／ユニーク数：', test3_1['選手名'].nunique())
print('test3_2件数:', test3_2['選手名'].count(), '／ユニーク数：', test3_2['選手名'].nunique())
````

````python
# ANSWER
# test3_1 で選手名が重複しているデータを確認
# DF.duplicated()は研修範囲外なので、確認方法はCSVファイルを確認する または 選手名でソートして確認する等
test3_1[test3_1.duplicated(subset='選手名', keep=False)].sort_values(by='選手名')
````

````python
# ANSWER
# 重複削除
test3_1.drop_duplicates(inplace=True)

# 重複削除結果確認
print('test3_1件数:', test3_1['選手名'].count(), '／ユニーク数：', test3_1['選手名'].nunique())
````

````python
# ANSWER
# test3_2 で選手名が重複しているデータを確認
# DF.duplicated()は研修範囲外なので、確認方法はCSVファイルを確認する または 選手名でソートして確認する等
test3_2[test3_2.duplicated(subset='選手名', keep=False)].sort_values(by='選手名')
````

````python
# ANSWER
# 2013年のみに絞る
test3_2 = test3_2.query('年 == 2013')

# 不要データ削除結果確認
print('test3_2件数:', test3_2['選手名'].count(), '／ユニーク数：', test3_2['選手名'].nunique())
````

````python
# ANSWER（コード変更なし）
# データ結合
test3_merge = pd.merge(test3_1, test3_2, on='選手名', how='inner')

# 年俸の差分を算出（2013年年俸 - 2012年年俸）
test3_merge['年俸差分'] = test3_merge['年俸_y'] - test3_merge['年俸_x']

# 年俸の差分の平均を算出
print(f"平均年俸差分：{test3_merge['年俸差分'].mean()} 万円")
````

````python
# ANSWER
# 結合後件数確認
print('結合後件数:', test3_merge['選手名'].count(), '／ユニーク数：', test3_merge['選手名'].nunique())
````
