#### 【Question2】現データを残しつつマスタデータの更新をする
---

'testdata2.csv'はある文房具店の商品と価格の入ったマスタデータです。  
来月から販売する商品に以下の変更が入る予定です。  

* 「研磨式消しゴム」の価格を高すぎと判断して500円 -> 450円に変更
* 「ただの輪ゴム」の商品名がタダじゃないとクレームがあった為、「何の変哲もない輪ゴム」に変更
* 「片手で書けるえんぴつ」は当たり前すぎて売れない為、販売終了
* 以前人気だった「曲がらないコンパス」を復刻版として新たに200円で販売

来月に向けてマスタデータを事前に変更することにしました。  
現商品マスタは現在使用中かつ過去分を残しておきたいので、直接修正は行わず一旦新しいデータフレームで修正し、  
来月になったら入れ替えを行うようにします。  
  
修正を行ったところ、現在使用しているマスタデータにも一部のみ変更が反映されてしまいました。  


以下のコードを修正し、どのコードをなぜ修正したか纏めてみましょう。

```python
# ライブラリインポート
import pandas as pd
test2 = pd.read_csv('work\doc\\02_Pythonデータ加工研修\\02_総合演習\data\\testdata2.csv', index_col=0)

# 読み込んだデータの表示
print(test2)

test2_new = test2.copy()#新しいオブジェクトを参照

# 価格・商品名の変更
test2_new.loc[test2_new['商品名']=='研磨式消しゴム', '価格'] = '450円'#「研磨式消しゴム」の価格を高すぎと判断して500円 -> 450円に変更
test2_new.loc[test2_new['商品名']=='ただの輪ゴム', '商品名'] = '何の変哲もない輪ゴム'

# 販売終了商品の削除
test2_new = test2_new.drop(index='S0073')

# 新商品の追加
newitem = pd.Series(['曲がらないコンパス', '200円'], index=test2_new.columns, name='S0003')
test2_new = test2_new.append(newitem)

# 修正の確認
print('')
print('=新商品マスタ==========')
print(test2_new)
print('')
print('=現商品マスタ==========')
print(test2)
```

●修正箇所と修正理由● ※なぜ想定外な結果となったかやどの箇所をどう修正したか等を簡単に整理して記載してみましょう（自由記述） 

修正前：test2_new = test2
修正後：test2_new = test2.copy()

#### 【Question3】野球選手の年俸の差額を確認する
---

'testdata3_1.csv'と'testdata3_2.csv'は、ヤクルトの野球選手のデータです。  

* 'testdata3_1.csv'：2012年の選手の属性情報と年俸および打率などの成績のデータ   
* 'testdata3_2.csv'：2013年の選手の年俸のデータ 

野球選手の年俸について分析することにしました。  
最初に2012年から2013年でどれくらい年俸に差があるのか差額の平均を算出しようと思います。  

データが2ファイルに分かれているので、まずデータを結合する必要があります。  
背番号は異なる選手で同じ背番号が存在しているので、選手名で結合することにしました。  
また年俸の差額を見たいので、2012年にも2013年にも在籍している選手のみを算出対象にします。  

実際に結合を行ったところ、結合後データのデータ数が2つの元データより増えてしまいました。  
ヤクルトで2012年と2013年の両方で在籍していた選手は**31人**のはずです。  

以下のコードを修正し、どのコードをなぜ修正したか纏めてみましょう。  
また今回は原因を探すための探索過程のコードも消さずに残しておきましょう。

```python
# ライブラリインポート
import pandas as pd
test3_1 = pd.read_csv('work\doc\\02_Pythonデータ加工研修\\02_総合演習\data\\testdata3_1.csv')
test3_2 = pd.read_csv('work\doc\\02_Pythonデータ加工研修\\02_総合演習\data\\testdata3_2.csv')
```
